# PySpark RDD - Movie Data Analytics
## Demo: Initialize SparkSession and SparkContext

```python
from pyspark.sql import SparkSession

# Initialize SparkSession
spark_session = SparkSession.builder \
                            .master("local") \
                            .appName("Colab PySpark") \
                            .getOrCreate()

print(spark_session)
```

```python
# Extract SparkContext from SparkSession
spark_context = spark_session.sparkContext

print(spark_context)
```

## Load Movie Data

```python
# Load raw text file into RDD
movie_raw_rdd = spark_context.textFile("/content/sample_data/movie_data.txt")

# Extract header row
header_row = movie_raw_rdd.first()
print("Headers:", header_row)

# Remove header row to get clean data RDD
movie_data_rdd = movie_raw_rdd.filter(lambda row: row != header_row)

movie_data_rdd.collect()
```

## Map Transformations
### Question 1: Extract Movie Names

```python
'''
Question 1: Extract Movie Names

Task:
- Extract only movie names from each record using map().

Expected Output:
- List of movie name strings
'''

# Extract movie name (index 1) from each record
movie_name_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                               .map(lambda movie_fields: movie_fields[1])

movie_name_rdd.collect()
```

### Question 2: Convert Movie Names to Uppercase

```python
'''
Question 2: Convert Movie Names to Uppercase

Task:
- Convert all movie names to uppercase using map().

Expected Output:
- List of movie names in UPPERCASE
'''

# Extract and convert movie name to uppercase
movie_name_uppercase_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                         .map(lambda movie_fields: movie_fields[1].upper())

movie_name_uppercase_rdd.collect()
```

### Question 3: Convert Movie Names to Lowercase

```python
'''
Question 3: Convert Movie Names to Lowercase

Task:
- Convert all movie names to lowercase using map().

Expected Output:
- List of movie names in lowercase
'''

# Extract and convert movie name to lowercase
movie_name_lowercase_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                         .map(lambda movie_fields: movie_fields[1].lower())

movie_name_lowercase_rdd.collect()
```

### Question 4: Extract Release Year as Integer

```python
'''
Question 4: Extract Release Year as Integer

Task:
- Extract only the release year and convert it to integer.

Expected Output:
- List of release years as integers
'''

# Extract release year (index 3) and cast to integer
movie_release_year_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                       .map(lambda movie_fields: int(movie_fields[3]))

movie_release_year_rdd.collect()
```

### Question 5: Key-Value Pair (Movie Name, Rating)

```python
'''
Question 5: Key-Value Pair (Movie Name, Rating)

Task:
- Create a key-value pair RDD: (MovieName, Rating).

Expected Output:
- List of (movie_name, rating) tuples
'''

# Create (movie_name, rating) key-value pairs
movie_name_rating_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                      .map(lambda movie_fields:
                                           (movie_fields[1], float(movie_fields[5]))
                                          )

movie_name_rating_rdd.collect()
```

### Question 6: Append '_MOVIE' to Every Movie Name

```python
'''
Question 6: Append "_MOVIE" to Every Movie Name

Task:
- Append the suffix "_MOVIE" to every movie name.

Expected Output:
- List of movie names with "_MOVIE" appended
'''

# Append "_MOVIE" suffix to each movie name
movie_name_suffixed_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                        .map(lambda movie_fields: movie_fields[1] + "_MOVIE")

movie_name_suffixed_rdd.collect()
```

### Question 7: Key-Value Pair (Genre, 1)

```python
'''
Question 7: Key-Value Pair (Genre, 1)

Task:
- Create a key-value pair RDD: (Genre, 1).

Expected Output:
- List of (genre, 1) tuples
'''

# Create (genre, 1) key-value pairs
movie_genre_count_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                      .map(lambda movie_fields: (movie_fields[2], 1))

movie_genre_count_rdd.collect()
```

### Question 8: Multiply Each Movie Duration by 2

```python
'''
Question 8: Multiply Each Movie Duration by 2

Task:
- Multiply each movie duration (in minutes) by 2.

Expected Output:
- List of doubled duration values
'''

# Extract duration (index 4), cast to int, and multiply by 2
movie_duration_doubled_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                           .map(lambda movie_fields: int(movie_fields[4]) * 2)

movie_duration_doubled_rdd.collect()
```

### Question 9: Key-Value Pair (Country, Rating)

```python
'''
Question 9: Key-Value Pair (Country, Rating)

Task:
- Create a key-value pair RDD: (Country, Rating).

Expected Output:
- List of (country, rating) tuples
'''

# Create (country, rating) key-value pairs
movie_country_rating_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                         .map(lambda movie_fields:
                                              (movie_fields[6], float(movie_fields[5]))
                                             )

movie_country_rating_rdd.collect()
```

### Question 10: Add 1.0 to Each Movie Rating

```python
'''
Question 10: Add 1.0 to Each Movie Rating

Task:
- Add 1.0 to each movie rating.

Expected Output:
- List of incremented rating values
'''

# Extract rating (index 5), cast to float, and add 1.0
movie_rating_incremented_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                             .map(lambda movie_fields: float(movie_fields[5]) + 1.0)

movie_rating_incremented_rdd.collect()
```

## Filter Transformations
### Question 11: Filter Movies with Rating > 8.5

```python
'''
Question 11: Filter Movies with Rating > 8.5

Task:
- Filter and return only movies where rating is greater than 8.5.

Expected Output:
- List of (movie_name, rating) tuples where rating > 8.5
'''

# Filter movies with rating greater than 8.5
high_rated_movies_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                      .map(lambda movie_fields: (movie_fields[1], float(movie_fields[5]))) \
                                      .filter(lambda movie_rating_pair: movie_rating_pair[1] > 8.5)

high_rated_movies_rdd.collect()
```

### Question 12: Filter Movies from 'Action' Genre

```python
'''
Question 12: Filter Movies from "Action" Genre

Task:
- Filter and return only movies belonging to the Action genre.

Expected Output:
- List of (movie_name, genre) tuples where genre == "Action"
'''

# Filter movies with genre equal to "Action"
action_genre_movies_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                        .map(lambda movie_fields: (movie_fields[1], movie_fields[2])) \
                                        .filter(lambda movie_genre_pair: movie_genre_pair[1] == "Action")

action_genre_movies_rdd.collect()
```

### Question 13: Filter Movies Released After 2015

```python
'''
Question 13: Filter Movies Released After 2015

Task:
- Filter and return only movies released after the year 2015.

Expected Output:
- List of (movie_name, release_year) tuples where release_year > 2015
'''

# Filter movies released after 2015
movies_after_2015_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                      .map(lambda movie_fields: (movie_fields[1], int(movie_fields[3]))) \
                                      .filter(lambda movie_year_pair: movie_year_pair[1] > 2015)

movies_after_2015_rdd.collect()
```

### Question 14: Filter Movies with Duration > 150 Minutes

```python
'''
Question 14: Filter Movies with Duration > 150 Minutes

Task:
- Filter and return only movies with duration greater than 150 minutes.

Expected Output:
- List of (movie_name, duration) tuples where duration > 150
'''

# Filter movies with duration greater than 150 minutes
movies_duration_over_150_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                             .map(lambda movie_fields: (movie_fields[1], int(movie_fields[4]))) \
                                             .filter(lambda movie_duration_pair: movie_duration_pair[1] > 150)

movies_duration_over_150_rdd.collect()
```

### Question 15: Filter Movies from 'India'

```python
'''
Question 15: Filter Movies from "India"

Task:
- Filter and return only movies produced in India.

Expected Output:
- List of (movie_name, country) tuples where country == "India"
'''
# Filter movies where country is India
indian_movies_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                  .map(lambda movie_fields: (movie_fields[1], movie_fields[6])) \
                                  .filter(lambda movie_country_pair: movie_country_pair[1] == "India")

indian_movies_rdd.collect()
```

### Question 16: Filter Movie Names Starting with 'T'

```python
'''
Question 16: Filter Movie Names Starting with "T"

Task:
- Filter and return only movie names that start with the letter "T".

Expected Output:
- List of movie name strings starting with "T"
'''
# Filter movie names starting with letter "T"
movies_starting_with_t_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                           .map(lambda movie_fields: movie_fields[1]) \
                                           .filter(lambda movie_name: movie_name.startswith("T"))

movies_starting_with_t_rdd.collect()
```

### Question 17: Filter Movies with Rating Between 8.0 and 8.5

```python
'''
Question 17: Filter Movies with Rating Between 8.0 and 8.5

Task:
- Filter and return movies with rating between 8.0 and 8.5 (inclusive).

Expected Output:
- List of (movie_name, rating) tuples where 8.0 <= rating <= 8.5
'''

# Filter movies with rating in the range [8.0, 8.5]
movies_rating_range_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                        .map(lambda movie_fields: (movie_fields[1], float(movie_fields[5]))) \
                                        .filter(lambda movie_rating_pair: 8.0 <= movie_rating_pair[1] <= 8.5)

movies_rating_range_rdd.collect()
```

### Question 18: Filter Movies from 'Animation' Genre

```python
'''
Question 18: Filter Movies from "Animation" Genre

Task:
- Filter and return only movies belonging to the Animation genre.

Expected Output:
- List of (movie_name, genre) tuples where genre == "Animation"
'''
# Filter movies with genre equal to "Animation"
animation_genre_movies_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                           .map(lambda movie_fields: (movie_fields[1], movie_fields[2])) \
                                           .filter(lambda movie_genre_pair: movie_genre_pair[1] == "Animation")

animation_genre_movies_rdd.collect()
```

### Question 19: Filter Movies with Duration < 100 Minutes

```python
'''
Question 19: Filter Movies with Duration < 100 Minutes

Task:
- Filter and return only movies with duration less than 100 minutes.

Expected Output:
- List of (movie_name, duration) tuples where duration < 100
'''

# Filter movies with duration less than 100 minutes
movies_duration_under_100_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                              .map(lambda movie_fields: (movie_fields[1], int(movie_fields[4]))) \
                                              .filter(lambda movie_duration_pair: movie_duration_pair[1] < 100)

movies_duration_under_100_rdd.collect()
```

### Question 20: Filter Movies Where Name Length > 10 Characters

```python
'''
Question 20: Filter Movies Where Name Length > 10 Characters

Task:
- Filter and return only movies whose name length exceeds 10 characters.

Expected Output:
- List of movie name strings with length > 10
'''
# Filter movies whose name has more than 10 characters
long_title_movies_rdd = movie_data_rdd.map(lambda movie_record: movie_record.split("|")) \
                                      .map(lambda movie_fields: movie_fields[1]) \
                                      .filter(lambda movie_name: len(movie_name) > 10)

long_title_movies_rdd.collect()
```
